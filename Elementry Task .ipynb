**Step 6:Identify and handle missing values**

```python
 #Check for missing values 
missing_values = df.isnull().sum()
```

```python
#Display the columns with missing values 
print(missing_values[missing_values > 0])
```

```python
# Fill missing values for numerical columns with median
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
# Fill missing values for categorical columns with mode
cat_cols = df.select_dtypes(include='object').columns
df[cat_cols] = df[cat_cols].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
#Now check for missing values 
df.isnull().sum()
```

```python
df.head(50)
```

**Step 7:Handle the Outliers and removing outliers**

```python
#Identify the outliers using IQR method 
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Remove outliers
df = df[~((df[num_cols] < (Q1 - 1.5 * IQR)) |(df[num_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df.head(100)
```

**Step 8:Correct Inconsistencies and Standardize Data Formats**

```python
# Standardize date format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Convert categorical variables to lowercase
df[cat_cols] = df[cat_cols].astype(str).apply(lambda x: x.str.lower())
```

```python
# Assuming the correct columns are 'Yards.Gained' and 'Yards.Net', replace these with the actual column names if different
if 'Yards.Gained' in df.columns and 'ydsnet' in df.columns:
    df['yardage_ratio'] = df['Yards.Gained'] / df['ydsnet']

if 'Yards.Gained' in df.columns:
    df['is_successful_play'] = df['Yards.Gained'] >= 10
```

**Step 11:Apply Transformation Techniques**

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# Replace infinite values with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
# Handle missing values: fill numerical columns with median and categorical with mode
num_cols = df.select_dtypes(include='number').columns
cat_cols = df.select_dtypes(include='object').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
df[cat_cols] = df[cat_cols].apply(lambda x: x.fillna(x.mode()[0]))
```

**Step 12:Apply scaling and normalization:**

```python
# Standardize numerical features
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
# Normalize 'yardage_ratio' using MinMaxScaler
min_max_scaler = MinMaxScaler()
df['yardage_ratio'] = min_max_scaler.fit_transform(df[['yardage_ratio']])
```
